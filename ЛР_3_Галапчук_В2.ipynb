```python
#Лабораторна робота 3
#Відокремлення коренів
import numpy as np

def f(x):
    return 3*x**4 + 4*x**3 - 12*x**2 - 5


def find_segments(): #відокремлюємо корені
    search_range = np.arange(-5, 5, 1)

    a = None
    previous_x = None
    current_x  = None
    segments = []

    for x in search_range:
        x = round(x, 4)
        current_x = f(x)
        if previous_x != None and previous_x * current_x < 0:
            segments.append((a, x))
        a = x
        previous_x = current_x
    return segments

segments = find_segments()
for a, b in segments:
    print(f'Found segment:  [{a}, {b}]')
```

```python
# Визначення функції та її похідної
def f(x):
    return 3*x**4 + 4*x**3 - 12*x**2 - 5

def df(x):
    return 12*x**3 + 12*x**2 - 24*x

def newton_method(x0, tol=1e-6, max_iter=100):

    print(f"Пошук кореня з початковим наближенням x0 = {x0}")
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)

        # Перевірка на ділення на нуль
        if dfx == 0:
            print("Похідна дорівнює нулю. Метод не може бути застосований.")
            return None

        # Формула методу Ньютона
        x_new = x - fx / dfx

        # Виведення інформації про ітерацію
        # Використовуємо науковий формат %.2e для маленьких чисел
        print(f"Ітерація {i+1}: x = {x_new:.6f}, |x_new - x_old| = {abs(x_new - x):.2e}")

        # Умови зупинки: або значення функції близьке до нуля,
        # або зміна між ітераціями дуже мала
        if abs(x_new - x) < tol or abs(fx) < tol:
            print(f"Знайдено корінь: {x_new:.6f} за {i+1} ітерацій.\n")
            return x_new

        x = x_new

    print(f"Досягнуто максимальної кількості ітерацій ({max_iter}).\n")
    return x

# --- ПРАВИЛЬНИЙ ВИКЛИК ФУНКЦІЇ ---

# Шукаємо корінь, починаючи з точки 1.5 (всередині вашого інтервалу [1, 2])
root1 = newton_method(2)

# Шукаємо корінь, починаючи з точки -2.5 (всередині вашого інтервалу [-3, -2])
root2 = newton_method(-3)
```

```python
def f(x):
    return 3*x**4 + 4*x**3 - 12*x**2 - 5

def df(x):
    return 12*x**3 + 12*x**2 - 24*x

def d2f(x):
    return 36*x**2 + 24*x - 24

def combined_method(a, b, tol=1e-6, max_iter=100):
    """
    Реалізація комбінованого методу.
    """
    print(f"Пошук кореня на інтервалі [{a}, {b}]")
    if f(a) * f(b) >= 0:
        print("f(a) * f(b) >= 0. Метод не застосовний.")
        return None

    # Визначення, який кінець оновлювати яким методом
    if f(a) * d2f(a) > 0:
        a_n, b_n = a, b # a - Ньютон, b - хорди
    else:
        a_n, b_n = b, a # b - Ньютон, a - хорди

    for i in range(max_iter):
        if abs(a_n - b_n) < tol:
            root = (a_n + b_n) / 2
            print(f"Знайдено корінь: {root:.6f} за {i} ітерацій.\n")
            return root

        # Зберігаємо попередні значення
        a_prev, b_prev = a_n, b_n

        # Наближення методом Ньютона
        a_n = a_prev - f(a_prev) / df(a_prev)

        # Наближення методом хорд
        b_n = b_prev - f(b_prev) * (b_prev - a_prev) / (f(b_prev) - f(a_prev))

        print(f"Ітерація {i+1}: a_n = {a_n:.6f}, b_n = {b_n:.6f}, |a_n-b_n|={abs(a_n-b_n):.2e}")


    print("Досягнуто максимальної кількості ітерацій.\n")
    return (a_n + b_n) / 2

# Пошук коренів на заданих інтервалах
root1 = combined_method(1, 2)
root2 = combined_method(-3, -2)
```
