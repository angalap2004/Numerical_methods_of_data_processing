```python
x0 = 2.9
y0 = 0.9
tolerance = 0.001

def f1(y):
    return 3 + np.cos(y)

def f2(x):
    return 0.5 - np.cos(x - 1)

def simple_iteration(x_start, y_start, e):
    xn, yn = x_start, y_start
    iteration_count = 0

    while True:
        iteration_count += 1
        xn1 = f1(yn)
        yn1 = f2(xn)

        if abs(xn1 - xn) < e and abs(yn1 - yn) < e:
            break

        xn, yn = xn1, yn1

    print("--- Метод простої ітерації ---")
    print(f"Розв'язок: x = {xn1:.5f}, y = {yn1:.5f}")
    print(f"Кількість ітерацій: {iteration_count}")

simple_iteration(x0, y0, tolerance)

def check_equations(vars):
    x, y = vars
    eq1 = np.cos(x - 1) + y - 0.5
    eq2 = x - np.cos(y) - 3
    return [eq1, eq2]

solution = optimize.root(check_equations, [0., 0.], method='hybr')

print("\n--- Перевірка за допомогою SciPy ---")
print(f"Точний розв'язок: x = {solution.x[0]:.5f}, y = {solution.x[1]:.5f}")
```
