Екзаменційний білет номер 2 Галапчук Андрій ФІТ 3-2
Завдання 3

```python
import math

def phi(x):
  return (x**3+4)/10

a=0.0
b=1.0
x_prev=0.5
eps=0.01
max_iter=100

for i in range(max_iter):
  x_next=phi(x_prev)

  if abs(x_next-x_prev)<eps:
      break

  x_prev=x_next

print("Метод простої ітерації")
print("Відрізок [{}, {}]".format(a,b))
print("Точність:", eps)
print("Кількість ітерацій:", i+1)
print("Наближене значення кореня:", round(x_next, 2))
```
