```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand

x=np.array([-1,0,2,3],dtype=float)
y=np.array([-16,-7,-1,20],dtype=float)

x1 = -0.5
x2 = 0.5
x3 = 1.5
x4 = 2.5

def lagrange_interpolation(x,y,x_test):
  n = len(x)
  p = np.zeros(n)
  for i in range(n):
    p_i = 1
    for j in range(n):
      if i != j:
        p_i *= (x_test-x[j])/(x[i]-x[j])
    p[i]=p_i
  return np.dot(y,p)


f1 = lagrange_interpolation(x,y,x1)
print("значення функції у точці x1 = ",f1.round(4))

f2 = lagrange_interpolation(x,y,x2)
print("значення функції у точці x2 = ",f2.round(4))


f3 = lagrange_interpolation(x,y,x3)
print("значення функції у точці x3 = ",f3.round(4))

f4 = lagrange_interpolation(x,y,x4)
print("значення функції у точці x4 = ",f4.round(4))

print("\n--- Побудова графіка (Графік 1) ---")
xnew = np.linspace(np.min(x), np.max(x), 100)
ynew = [lagrange_interpolation(x, y, i) for i in xnew]

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o', label='Вузлові точки (Варіант 2)', markersize=8)
plt.plot(xnew, ynew, label='Багаточлен Лагранжа $L_3(x)$')

plt.plot(x1, f1, 'ro', label='Розраховані точки')
plt.plot(x2, f2, 'ro')
plt.plot(x3, f3, 'ro')
plt.plot(x4, f4, 'ro')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Polynomial (Варіант 2)')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Перевірка (Графік 2) ---")
f = lagrange_interpolation(x,y)

print(f"Багаточлен (Scipy): \n{f}")

fig = plt.figure(figsize = (7,5))
plt.plot(xnew, f(xnew), 'b', label='Scipy $L_3(x)$')
plt.plot(x, y, 'ro', label='Вузлові точки')

plt.plot(x1, f1, 'o', label='Розраховані точки')
plt.plot(x2, f2, 'o')
plt.plot(x3, f3, 'o')
plt.plot(x4, f4, 'o')

plt.title('Lagrange Polynomial (Перевірка Scipy)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
